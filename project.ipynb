#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
# TODO
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('assets/real_estate.csv', sep=';')
south_belt_pops = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
south_belt_data = ds[ds['level5'].isin(south_belt_pops)].copy()

def normalize_prices(group):
    min_price = group.min()
    max_price = group.max()
    return (group - min_price) / (max_price - min_price)

south_belt_data['normalized_price'] = south_belt_data.groupby('level5')['price'].transform(normalize_prices)  # 'price' used directly

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

for i, pop in enumerate(south_belt_pops):
    row = i // 2
    col = i % 2
    population_data = south_belt_data[south_belt_data['level5'] == pop]['normalized_price']
    axes[row, col].hist(population_data, alpha=0.7, label=pop)
    axes[row, col].set(title=f'Normalized Price Distribution - {pop}', xlabel='Normalized Price (0-1)', ylabel='Frequency')
    axes[row, col].legend()

plt.tight_layout()
plt.show()
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph. You must use the subset obtained in Exercise 14 (★★☆)
 
Hint: Make a scatter plot of each population using subplots.

```python
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('assets/real_estate.csv', sep=';')
south_belt_pops = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
south_belt_data = ds[ds['level5'].isin(south_belt_pops)].copy()

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

for i, pop in enumerate(south_belt_pops):
    row = i // 2
    col = i % 2

    city_data = south_belt_data[south_belt_data['level5'] == pop]

    if not city_data.empty:
    
        axes[row, col].scatter(city_data['rooms'], city_data['price'], label=f"{pop} (Rooms vs. Price)", s=10, alpha=0.7)

        axes[row, col].scatter(city_data['bathrooms'], city_data['price'], label=f"{pop} (Bathrooms vs. Price)", s=10, alpha=0.7)

        axes[row, col].set_title(f'Property Features vs. Price in {pop}')
        axes[row, col].set_xlabel('Number of Rooms/Bathrooms')
        axes[row, col].set_ylabel('Price')
        axes[row, col].legend()

plt.tight_layout()
plt.show()
```
